```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load dataset
df = pd.read_csv('us_air_pollution_2012_2021.csv')  # Upload this file in Colab
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Select time series
ts = df['PM2.5 (µg/m³)'].dropna()

# Decompose using additive model
decomposition = seasonal_decompose(ts, model='additive', period=365)

# Plot components
decomposition.plot()
plt.suptitle('Time Series Decomposition of PM2.5', fontsize=16)
plt.tight_layout()
plt.show()
```
